```python
Data = pd.read_csv(r'G:\data_angle.csv')
y = Data.iloc[:,16:]
X = Data.iloc[:,:16]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(Data, test_size=0.2, random_state=42)
```

```python
train_set
```

```python
Data = train_set.drop("RightorLeft", axis=1) # drop labels for training set
pre_label = train_set["RightorLeft"].copy()
```

```python

from sklearn.ensemble import RandomForestClassifier
classifier_ran = RandomForestClassifier(n_estimators=100, max_depth=12, random_state=42)
classifier_ran.fit(Data,pre_label)
classifier_ran.score(X,y)*100
```

```python
from sklearn.linear_model import LogisticRegression
classifier_log = LogisticRegression(random_state=42)
classifier_log.fit(Data,pre_label)
classifier_log.score(X,y)*100
```

```python
from sklearn import svm
SVM = svm.LinearSVC()
SVM.fit(Data,pre_label)

SVM.score(X,y)*100
```

```python
from sklearn.neural_network import MLPClassifier
NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=42)
NN.fit(Data,pre_label)
```
